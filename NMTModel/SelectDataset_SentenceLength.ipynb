```python
import re
from pickle import dump
from unicodedata import normalize
from numpy import array
import string
import codecs
import numpy as np

import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
from gensim.models.keyedvectors import KeyedVectors

from pickle import load
from numpy.random import rand
from numpy.random import shuffle
 
# load a dataset
def load_sentences(dataFileName):
    with codecs.open(dataFileName, 'r', 'utf-8') as fin:
        sentences = fin.readlines()
    return sentences
```

```python
# max sentence length
def max_length(lines) :
    return len(lines.split())
```

```python
en_sentences = load_sentences('train_en.txt')
sen_sentences = load_sentences('train_sen.txt')
```

```python
def norm_unicode(line) :
    line = normalize('NFD', line).encode('ascii', 'ignore')
    line = line.decode('UTF-8')
    return line
```

```python
iter = 0


MinVal  = 5
MaxVal  = 8


RowCount = -1
edit_en_sentences = []
edit_sen_sentences = []
en_file = open("en_5_8temp.txt", "w")
sen_file = open("sen_5_8temp.txt", "w")
for lines in en_sentences :
    RowCount = RowCount +1
    if max_length(lines) > MinVal :
        if max_length(lines) < MaxVal :
            if max_length(sen_sentences[RowCount]) > MinVal :
                if max_length(sen_sentences[RowCount]) < MaxVal :
                    iter = iter +1
                    print(iter) 
                    print('Complex Sentence : %s' %lines)
                    print('Simple Sentence : %s' %sen_sentences[RowCount])
                    en_file.write(norm_unicode(en_sentences[RowCount]))
                    sen_file.write(norm_unicode(sen_sentences[RowCount]))
                    #edit_en_sentences.append(en_sentences[RowCount])
                    #edit_sen_sentences.append(sen_sentences[RowCount])
                    print('----\n')
en_file.close()
sen_file.close()
```
